```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# do this in anaconda navigator: $ pip install plotly
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
plotly.offline.init_notebook_mode(connected=True)
```

```python
ndaqTr212 = pd.read_csv("NDAQ100_trading212.csv", header=0, parse_dates=['Date']) # , index_col='Date'
ndaqTr212 = ndaqTr212.set_index(ndaqTr212['Date'])
```

```python
ndaqTr212.head()
```

```python
# Force Index = Volume(today) Â· (Close(today) - Close(yesterday))
tradingDays = ndaqTr212.shape[0]
close = ndaqTr212.loc[:,"Close"]
closeV = close.values
volume = ndaqTr212.loc[:,"Volume"]
volumeV = volume.values
forceIndex = []
EMA = []
K = 2/(2+1) # We only use 2 Days as the window
for i in range(1,tradingDays):
    #print(close[i], close[i-1])
    tmp = volumeV[i] * (closeV[i] - closeV[i-1])
    forceIndex.append(tmp)
EMAseed = (forceIndex[1] + forceIndex[0])/2
EMA.append(EMAseed)
for e in range(2,len(forceIndex)):  
    #EMAstart = forceIndex[e] * K + EMAseed * (1 - K)
    EMAtoday = forceIndex[e] * K + EMA[e-2] * (1 - K)
    EMA.append(EMAtoday)
```

```python
figEx = plt.figure(figsize=(20,10))
ax = figEx.add_subplot(111)
ax.set_title('BUY/SELL during 2018-2019')
ax.axhline(y=0, color='r')
plt.plot(np.arange(len(EMA)), EMA, label='EMA')
plt.xlabel("Days")
plt.ylabel("Slope")
ax.legend(loc='best')
plt.show()
```
