```python
# define the cost function in a vectorized way
def computeCost(X, y, theta):
    m = len(y)
    J = 0
    # In the course and the formula we multiply theta.transpose x X. But in a vectorized way the following works:
    # Here we multiply a matrix of 97x2  2x1. The result will be a vector 'h' of dimensions 97x1.
    h = X.dot(theta)
    # 'h' and 'y' have the same dimensions. Calculate the square error.
    sq_er = (h-y)**2
    # Calculate Cost Function and return it.
    J = 1/(2*m) * np.sum(sq_er)
    return J
```

```python
# define the Gradient Descent
def gradientDescent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = []
    
    for i in range(num_iters):
        # Here we multiply a matrix of 97x2  2x1. The result will be a vector 'h' of dimensions 97x1.
        h = X.dot(theta)
        # 'X' has dimension 97x2 and '(h-y)' 97x1. To multiply, we need to transpose 'X' and we get dimension 2x97.
        # Will do a multiplication of 2x97  97x1. The result will have dims of 2x1.
        error = np.dot(X.transpose(),(h - y))
        # Complete the descent including 'aplha' and '1/m'
        descent = alpha * 1/m * error
        # Update the theta vector
        theta -= descent
        # Compute Cost Function
        J_history.append(computeCost(X,y,theta))
    return theta, J_history
```

```python
# ufunc 'subtract' output (typecode 'O') could not be coerced to provided output parameter (typecode 'd') according to the casting rule ''same_kind''
theta, J_history = gradientDescent(X, y, theta, alpha, iterations)
# Error because of distinct types
# Fix for X and y -> X = np.array(X, dtype=np.float64); y = np.array(y, dtype=np.float64)
theta
```

```python
# We have the DataFrame 'ndaq2018'
# Define here constant variables
lengthDf = len(ndaq2018.loc[:,'Open'])
alpha = 0.03
iterations = 500
window = 12
primer = [0 for i in range(window)]
close = ndaq2018.loc[:,'Close']
close12 = close.values
```

```python
# Define useful functions
def calc_slope(window, stockPrices, alpha, iterations):
    slope_history = []
    lengthDf = len(stockPrices.loc[:,'Open'])
    for i in range(lengthDf-window):
        nqTmp = stockPrices.iloc[i:window+i]
        l = [float(n) for n in range(window)]
        nqTmp.insert(1, 'Time', l)
        df = nqTmp.values
        m = len(df[:,1])
        X = np.append(np.ones((m,1)), df[:,1].reshape(m,1), axis=1)
        y = df[:,5].reshape(m,1)
        thetaTmp = np.zeros((2,1))
        X = np.array(X, dtype=np.float64)
        y = np.array(y, dtype=np.float64)
        thetaTmp, J_history = gradientDescent(X, y, thetaTmp, alpha, iterations)
        slope_history.append(thetaTmp[1])
    return slope_history

def generate_trades(slope_2019):
    positions_2019 = []
    for l in range(len(slope_2019)-4):
    #Reverse direction
    #for l in range(len(slope_2019)-1, -1, -1): 
        direction = ""
        if slope_2019[l] == 0:
            continue
        if (slope_2019[l] > (slope_2019[l+1])) and ((slope_2019[l+1]) > (slope_2019[l+2])) and ((slope_2019[l+2]) > (slope_2019[l+3])):
    #if (slope_2019[l-2] > (slope_2019[l-1])) and ((slope_2019[l-1]) > (slope_2019[l])):
            direction = "SELL"
            positions_2019.append(direction)
            #positions_2019.append(datePrices_2019[l])
        elif (slope_2019[l] < (slope_2019[l+1])) and ((slope_2019[l+1]) < (slope_2019[l+2])) and ((slope_2019[l+2]) < (slope_2019[l+3])):
    #elif (slope_2019[l-2] < (slope_2019[l-1])) and ((slope_2019[l-1]) < (slope_2019[l])):        
            direction = "BUY"           
            positions_2019.append(direction)
            #positions_2019.append(datePrices_2019[l])
        else:
            direction = "--"
            positions_2019.append(direction)
            #positions_2019.append(datePrices_2019[l])
    return positions_2019

def generate_trades_v2(slope_2019):
    positions_2019 = []
    for l in range(len(slope_2019)-2):
    #Reverse direction
    #for l in range(len(slope_2019)-1, -1, -1): 
        direction = ""
        if slope_2019[l] == 0:
            continue
        if (slope_2019[l] > (slope_2019[l+1])) and ((slope_2019[l+1]) > (slope_2019[l+2])):
    #if (slope_2019[l-2] > (slope_2019[l-1])) and ((slope_2019[l-1]) > (slope_2019[l])):
            direction = "SELL"
            positions_2019.append(direction)
            #positions_2019.append(datePrices_2019[l])
        else:
            direction = "BUY"           
            positions_2019.append(direction)
    return positions_2019

def calc_profit(direction, positions, prices, window, datePrices):
    trades = []
    profits = []
    trade = []
    dates = []
    date = []
    length = len(positions)
    for i, val in enumerate(positions):
        if i == length - 1:
            break
        if val == direction:
            trade.append(i+window+3)
            date.append(datePrices[i+window+3])
            if positions[i+1] != direction:
                trade.append(i+1+window+3)
                date.append(datePrices[i+1+window+3])
                trades.append(trade)
                dates.append(date)
                trade = []
                date = []
    for t in trades:
        profits.append(prices[t[-1]]-prices[t[0]]) if direction == 'BUY' else profits.append(prices[t[0]]-prices[t[-1]])
    return trades, profits, dates

def plot_profits(lengthDf, close_2019, openPrices_2019, buyTrades, sellTrades):
    
    figEx = plt.figure(figsize=(20,10))
    ax = figEx.add_subplot(111)
    ax.set_title('BUY/SELL during 2018-2019')
    #plt.plot(np.arange(lengthDf), slope12, label='slope')
    plt.plot(np.arange(lengthDf), close_2019, label='close')
    # because openPrices_2019 has bigger length, we have to trimm like close_2019
    tmp = openPrices_2019[0:lengthDf]
    plt.plot(np.arange(lengthDf), tmp, label='open')
    ymin, ymax = plt.ylim()
    #for d in buyPos:
    #    ax.axvline(x=d, color='g')
    #for d in sellPos:
    #    ax.axvline(x=d, color='r')
    for s in buyTrades:
        li = np.array(s)
        ax.fill_between(li, ymin, ymax, where=li>1, facecolor='green', alpha=0.5)
    for s in sellTrades:
        li = np.array(s)
        ax.fill_between(li, ymin, ymax, where=li>1, facecolor='red', alpha=0.5)
    plt.xlabel("Days")
    plt.ylabel("Slope")
    ax.legend(loc='best')
    plt.show()

def plot_price_vs_slope(days, closePrices, slopes, stockPrices, showDays=False):
    lengthDf = len(stockPrices.loc[:,'Open'])
    plt.figure()
    f, axes = plt.subplots(2, 1, figsize=(20, 10))
    if showDays:
        for d in days:
            axes[0].axvline(x=d, color='k')
            axes[1].axvline(x=d, color='c')
    axes[0].plot(np.arange(lengthDf), closePrices)
    axes[0].set_ylabel('close')

    axes[1].plot(np.arange(lengthDf), slopes, color='r')
    axes[1].set_ylabel('slope')
    plt.show()
```
